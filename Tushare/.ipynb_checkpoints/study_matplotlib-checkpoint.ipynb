```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 画数据
plt.plot([1,2,3,4],[2,3,4,5],'o-r',label='Line a')
plt.plot([1,2,3,4],[5,2,7,3],'^--b',label="Line b")
plt.title('matplotlib Test Plot')
plt.xlabel("x num")
plt.ylabel("y num")
plt.xlim(0,5)
plt.ylim(0,10)
plt.xticks([i for i in range(6) if i%2==0])
plt.yticks([i for i in range(10) if i%2==0])
plt.legend()
plt.show()
```

```python
# 画函数
x = np.linspace(-10,10,100)
a = x
b = x**2
plt.plot(x,a,'-g')
plt.plot(x,b,'--r')
plt.show()
```

```python
# 画布
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.plot([1,2,3,4],[2,5,3,6])
ax2 = fig.add_subplot(2,1,2)
ax2.plot([1,2,3,4],[4,5,2,1])
fig.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def poisson(k:float,l:float) :
    from math import exp,factorial
    return pow(l,k)*exp(-l)/factorial(k)

total = 100
premium = 1000
cost = 2000
p_max = math.trunc(total * premium / cost)
a = np.array([i*1.0 for i in range(50)])
b = []
for aa in a:
    b.append(poisson(aa,40.0))
plt.plot(a,b,'o-r',label='Poisson')
plt.show()
```
